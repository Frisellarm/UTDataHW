Three observable trends that I found after analyzing the math and reading scores data:
* In terms of school type, charter schools' scores were higher than district schools' scores in every category (with an overall % passing rate of 95.1% compared to 73.67% for district schools). This could be due to the curriculum of the charter schools and the knowledge level of the students that attend charter schools versus district schools.

* In terms of school size, small (less than 1000 students) and medium (1000-2000 students) size schools had nearly the same percentage of scores (around 95%) and were much higher than large size schools (greater than 2000 students) which had overall % passing rate of 76.36%. This could support the theory that much larger sized schools have more difficulty accomodating learning for students.

* In terms of spending ranges per student, the data showed that the highest amount of spending per student actually yielded the lowest % of passing test scores (73.6%) compared to spending range of less than $615 which had an overall passing rate of 95 percent. Charter schools tend to have lower number of students than district schools therefore charter schools spending range per student would be less than district schools.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results


* Optional: give the displayed data cleaner formatting

```python
#Total number of schools
total_schools = len(school_data_complete["school_name"].unique())
print(total_schools)

#Total number of students
total_students = len(school_data_complete["Student ID"].unique())
print(total_students)

#Total budget
total_budget_list = school_data_complete["budget"].unique()
total_budget = total_budget_list.sum()
print(total_budget)

#Average math score
average_math_score = school_data_complete["math_score"].mean()
print(average_math_score)

#Average reading score
average_reading_score = school_data_complete["reading_score"].mean()
print(average_reading_score)

#Overall passing rate
overall_passing_rate = (average_math_score + average_reading_score)/2
print(overall_passing_rate)

#Percent passing math
passing_math = school_data_complete["math_score"] >= 70.0
students_math = passing_math.value_counts()
num_students_passing_math = students_math.iloc[0][True][0]
percent_passing_math = (num_students_passing_math/total_students) * 100.0
print(percent_passing_math)

#Percent passing reading
passing_reading = school_data_complete["reading_score"] >= 70.0
students_reading = passing_reading.value_counts()
num_students_passing_reading = students_reading.iloc[0][True][0]
percent_passing_reading = (num_students_passing_reading/total_students) * 100.0
print(percent_passing_reading)
```

```python
#Create district summary dataframe
district_dict = {"Total Schools": total_schools, 
                 "Total Students": total_students,
                 "Total Budget": total_budget,
                 "Average Math Score": average_math_score,
                 "Average Reading Score": average_reading_score,
                 "% Passing Math": percent_passing_math,
                 "% Passing Reading": percent_passing_reading,
                 "% Overall Passing Rate": overall_passing_rate
                }
district_summary_df = pd.DataFrame(district_dict,index=[0])
#Format Total Budget with $ in front and to 2 digit decimal floating point
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

#Output final district summary
district_summary_df
```

```python
#Percent passing reading by school_name
school_name_percent_passing_reading = (num_students_reading_pass/school_name_total_students) * 100.0
print(school_name_percent_passing_reading)
```

```python
#Overall passing rate percent by school_name is average of percent_passing_math and percent_passing_reading
percent_overall_passing_rate = (school_name_percent_passing_math + school_name_percent_passing_reading)/2
print(percent_overall_passing_rate)
```

```python
school_type=school_data_complete.groupby("school_name")["type"].unique()
print(school_type)
print(type(school_type))
# remove the [ in front of District and Charter and ] at end of District and Charter
school_type_clean = school_type.str[0]
print(school_type_clean)
print(type(school_type_clean))
```
